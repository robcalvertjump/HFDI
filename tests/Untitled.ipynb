Data source: https://digital.nhs.uk/data-and-information/data-collections-and-data-sets/data-collections/general-practice-data-collections

```python
import geopandas as gpd
import fiona
import matplotlib.pyplot as plt

gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'
#df_catchment = gpd.read_file('data/eDecJan-Mar2020.kml', driver='KML')
df_catchment = gpd.read_file('data/2018_10_pcwt_boundaries.kml', driver='KML')
```

```python
df_catchment.head()
```

```python
len(set(df_catchment['Name']))
```

```python
df = df_catchment

fig, ax = plt.subplots(figsize=(10, 10), dpi=200)
df.plot(ax=ax, alpha=0.1)

plt.show()
```

```python
df_catchment['geometry'][0].area
```

```python
# Find the catchment areas that are faulty

mask = df_catchment.apply(lambda row: row['geometry'].area > 0.2, axis=1)
df_faulty = df_catchment[mask]
```

```python
df = df_faulty

fig, ax = plt.subplots(figsize=(10, 10), dpi=200)
df.plot(ax=ax)

plt.show()
```
